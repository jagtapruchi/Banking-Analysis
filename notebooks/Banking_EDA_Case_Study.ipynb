```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("/content/Banking.csv")
df.head()
```

```python
bins = [0,100000, 300000, float('inf')]  # ranges are: 0-1L, 1L-3L, 3l-infinity
labels = ['Low', 'Medium', 'High']
df['Income_Category'] = pd.cut(df['Estimated Income'], bins=bins, labels=labels)
```

```python
df['Income_Category'].value_counts()
```

```python
df['Income_Category'].value_counts().plot(kind='bar')
```

```python
# Examine the distribution of unique categories in categorical columns:

categorical_cols = ['BRId','GenderId','IAId','Amount of Credit Cards','Nationality','Occupation','Fee Structure','Loyalty Classification','Properties Owned','Risk Weighting','Income_Category']
for col in categorical_cols:
    print(f"Value counts for '{col}':")
    display(df[col].value_counts())
```

## Bivariate Analysis by GenderId

```python
# bar plot showing the count of each category in the column predictor, split by GenderId.

for i, predictor in enumerate(df[['BRId','IAId','Amount of Credit Cards','Nationality','Occupation','Fee Structure','Loyalty Classification','Properties Owned','Risk Weighting','Income_Category']]):
  plt.figure(i)
  sns.countplot(data=df,x=predictor,hue='GenderId')
```

## Bivariate Analysis by Nationality

```python
# bar plot showing the count of each category in the column predictor, split by Nationality.

for i, predictor in enumerate(df[['BRId','GenderId','IAId','Amount of Credit Cards','Occupation','Fee Structure','Loyalty Classification','Properties Owned','Risk Weighting','Income_Category']]):
  plt.figure(i)
  sns.countplot(data=df,x=predictor,hue='Nationality')
```

## Count plot of all categorical cols

```python
for col in categorical_cols:
    plt.figure(figsize=(8, 4))
    sns.countplot(data=df, x=col)
    plt.title(f'Count Plot of {col}')
    plt.xticks(rotation=45)  # X-label ticks
    plt.tight_layout()
    plt.show()
```

## Numerical Analysis

```python
numerical_cols = ['Age','Estimated Income','Superannuation Savings','Amount of Credit Cards','Credit Card Balance','Bank Loans','Bank Deposits', 'Checking Accounts', 'Saving Accounts', 'Foreign Currency Account', 'Business Lending', 'Properties Owned', 'Risk Weighting', 'BRId', 'GenderId', 'IAId']
plt.figure(figsize=(12,20))
for i,col in enumerate(numerical_cols):
  plt.subplot(8,2,i+1)
  sns.histplot(data=df[col],kde=True)
  plt.title(col)
plt.show()
```

## Heatmaps

```python
numerical_cols = ['Age','Estimated Income','Superannuation Savings','Amount of Credit Cards','Credit Card Balance','Bank Loans','Bank Deposits', 'Checking Accounts', 'Saving Accounts', 'Foreign Currency Account', 'Business Lending', 'Properties Owned', 'Risk Weighting', 'BRId', 'GenderId', 'IAId']

correlation_matrix = df[numerical_cols].corr()
plt.figure(figsize=(12,12))
sns.heatmap(correlation_matrix,annot=True,cmap='crest',fmt='.2f')
plt.title('Correlation Matrix')
```
